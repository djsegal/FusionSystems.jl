## The Future of Fusion Energy
Pulsed vs Steady-State Tokamaks


```julia
# author: dan segal (dan@seg.al)

using FusionSystems
using SimplePlots
```

---

### Steady State Tokamaks

```julia
used_fracs = 0.5:0.25:1.5

@demo for reactor in ["ARC", "Steady"], P_F=used_fracs, eta_CD=used_fracs, epsilon=used_fracs
  scan = Scan(reactor, P_F=P_F, eta_CD=eta_CD, epsilon=epsilon)
  FusionSystems.plot(scan, :B_C, :cost)
  
  xlims!(0,30) ; ylims!(0,150)
  title!("$(reactor) Cost Curves")
end
```

---

### Pulsed Tokamaks

```julia
used_fracs = 0.5:0.25:1.5

@demo for reactor in ["DEMO", "Pulsed"], P_F=used_fracs, N_G=used_fracs
  scan = Scan(reactor, P_F=P_F, N_G=N_G)
  FusionSystems.plot(scan, :B_C, :cost)
  
  xlims!(0,30) ; ylims!(0,150)
  title!("$(reactor) Cost Curves")
end
```
